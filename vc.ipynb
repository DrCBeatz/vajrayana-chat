```python
import bs4
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/Yongey_Mingyur_Rinpoche"

# Get the page
page = requests.get(url)
page = bs4.BeautifulSoup(page.content, 'html.parser')
text_content = page.get_text()
```

```python
cleaned_text = "\n".join(line.strip() for line in text_content.splitlines() if line.strip())
cleaned_text
```

```python
with open("mingyur_rinpoche_wikipedia.txt", "w", encoding="utf-8") as f:
    f.write(cleaned_text)
```
